## Exploration des données

Suite au nettoyayge des données nous nous retrouvons avec un dataset qui contient 13 variables et 200,137 observations. 

Parmis ces variables il y a 2 variables qui ont environ 1% de données manquantes (0.7 & 0.8) ce qui reste raisonable pour un dataset de cette taille. 

Parmis les différentes variables nous avons :

* type_emplacement ; arbre
* domanialite ; emplacement ou se situe l'arbre (ex. jardin, cimetière)
* arrondissement ; le nom de l'arrondissement ou se trouve l'arbre
* lieu ; endroit ou l'arbre se situe
* id_emplacement ; id de l'emplacement de l'arbre
* libelle_francais ; type d'arbre planté (ex. abricotier, alangium)
* genre ; genre de l'arbre (ex. pour érable on a acer)
* espece ; espèce de l'arbre (ex. pour érable on a negundo)
* circonference_cm ; circonference de l'arbre
* hauteur_m ; hauteur de l'arbre en mètre
* geo_point_2d_a ; coordonnées x
* geo_point_2d_b ; coordonnées y

On peut remarquer à l'aide du .info() que nous avons principalement de la données non int sous la forme 'object'.

### Moyennes, minimum, maximum et quartiles

```python
df.describe()
```

Ce qu'on peut observer avec ce describe c'est qu'il y a une très grande variation entre le maximum et le minimum pour la circonference et la hauteur. De la même manière on remarque qu'il y a un écart-type de 673 & 1971 pour la circonférence en cm et la hauteur en m des arbres. Cela signifie qu'il y a un ou plusieurs arbre qui sont beaucoup plus grands ou beaucoup plus petits que les autres ce qui augmente ces valeurs.

Au vu de la moyenne de ces 2 variables nous pouvons assumer qu'il s'agit d'une valeur bien plus grande que le reste des observations qui fait qu'il y a une variation aussi importante. Les quartiles confirment cette hypothèse car ils se rapprochent de la moyenne contrairement à la valeur maximale.

### Heatmap des corrélations

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Matrice de corrélation des variables')
plt.show()
```

Ce que nous remarquons avec cette heatmap c'est qu'il n'y a pas de forte corrélation entre les variables numériques. En effet, plus le chiffre se rapproche de 1 plus la corrélation est élevé mais ici nous avons une corrélation force de 0.3 mais elle existe entre deux variables qui n'ont pas de lien particulier. 

Etant donné que nous avons beaucoup plus de variables non numériques cela rend l'utilisation d'une matrice de corrélation difficile car il n'y a pas forcément d'élément de mesure mis à part l'utilisation d'un graphique de type histogramme ou bar plot afin d'identifier les différentes corrélations. 

Il faudrait explorer les arbres, leur genre et les emplacements afin de voir si on peut observer une corrélation entre ces différentes variables.  

### Histogramme de la circonférence en cm

```python
df['circonference_cm'].plot.hist(bins=np.logspace(0, 5, 20), edgecolor='black', log=True)

plt.xscale('log')
plt.xticks([1, 10, 100, 1000, 10000, 100000], ['1', '10', '100', '1k', '10k', '100k'])

plt.title('Distribution de la circonférence en cm')
plt.xlabel('Circonférence en cm (échelle logarithmique)')
plt.ylabel('Fréquence')
plt.show()
```

Ce graphique nous permet de voir la réprésentation des différentes circonférences. Ce que nous remarquons confirme notre observation précédente, il y a une grande disparité au niveau des circonférences avec certaines qui sont bien plus grandes que les autres.

### Barplot de la fréquence des domanialités 

```python
df['domanialite'].value_counts().plot.bar()
plt.title('Fréquence des domanialite')
plt.xlabel('Type d\'emplacement')
plt.ylabel('Fréquence')
plt.show()
```

Ici nous remarquons que la majorité des emplacement sont de type 'Alignement', c\'est à dire que la majorité des arbres sont des arbes sont alignés sur des routes/rues etc. Ensuite nous avons les jardins et les cimetières pour les types d'emplacements qui ont la plus grande quantité d'arbres. 

### Top 5 des genres (types) et espèces les plus présents

```python
plt.figure(figsize=(15, 5))
top_libelle_francais = df['genre'].value_counts().head(5)
sns.barplot(x=top_libelle_francais.index, y=top_libelle_francais.values)
plt.title('Top 5 des Genres Français les plus présents')
plt.xlabel('genre')
plt.ylabel('Fréquence')
plt.show()

plt.figure(figsize=(15, 5))
top_genre = df['espece'].value_counts().head(5)
sns.barplot(x=top_genre.index, y=top_genre.values)
plt.title('Top 5 des Especes les plus présents')
plt.xlabel('espece')
plt.ylabel('Fréquence')
plt.show()
```

Ici on remarque que le genre d'arbre le plus représenté est le platanus et que l'espèce la plus représentée et la x hispanica. 

### Barplot du nombre d'arbres par arrondissement

```python
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='arrondissement')
plt.title('Nombre d\'arbres par arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Nombre d\'arbres')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Ce graphique nous permet de voir quels arrondissement ont quelle quantité d'arbre. Par exemple on remarque que le 4ème arrondissement possède environ 2,500 arbres.
